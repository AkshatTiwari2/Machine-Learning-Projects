```python
titles = list(df.loc[:,'movie_title'])
```

```python
print(titles[:5])
```

```python
print(type(list(titles)))
freq = {}
print(freq.get(0))
```

```python
freq = {}

for title in titles:
  length = len(title)
  if freq.get(length) == None:
    freq[length] = 1
  else:
    freq[length] += 1
```

```python
print(freq)
```

```python
X = np.array(list(freq.keys()))
Y = np.array(list(freq.values()))
print(X,Y)
```

```python
plt.style.use('seaborn-v0_8')
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Frequency")
plt.show()
```
