```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples = 500,n_features=4,n_informative=4,noise=1.8,random_state=11)
X.shape,Y.shape
```

```python
X_D = pd.DataFrame(X)
X_D.describe()
```

```python
Y = Y.reshape((-1,1))
Y.shape
```

```python
# normalise
X = (X-X.mean())/X.std()
```

```python
# Visualise
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.scatter(X[:,i],Y)
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
```

```python
def predict(X,theta):
  return np.dot(X,theta)

def error(X,Y,theta):
  m = X.shape[0]
  y_ = predict(X,theta)
  error = np.sum((y_-Y)**2)
  return error/m

def getThetaClosedForm(X,Y):
  first = np.linalg.pinv(np.dot(X.T,X))
  second = np.dot(X.T,Y)
  theta = np.dot(first,second)
  e = error(X,Y,theta)
  return theta,e
```

```python
theta,error =  getThetaClosedForm(X,Y)
```
